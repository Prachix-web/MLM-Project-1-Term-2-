````python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd

from sklearn.impute import SimpleImputer

def CP00_impute_columns_with_mean_or_median(CP00_df, columns):
    for col in columns:
        # Ensure column is numeric
        CP00_df[col] = pd.to_numeric(CP00_df[col], errors='coerce')

        # Replace invalid values with NaN
        CP00_df[col].replace([np.inf, -np.inf], np.nan, inplace=True)

        # Skip if column has no missing values
        if CP00_df[col].isnull().sum() == 0:
            continue

        # Calculate mean and median
        CP00_col_mean = CP00_df[col].mean()
        CP00_col_median = CP00_df[col].median()

        # Choose strategy based on significant difference
        if abs(CP00_col_mean - CP00_col_median) / max(abs(CP00_col_mean), abs(CP00_col_median)) > 0.1:  # Significant difference
            print(f"Imputing '{col}' with median (significant difference between mean and median)")
            CP00_imputer = SimpleImputer(strategy='median')
        else:
            print(f"Imputing '{col}' with mean (no significant difference between mean and median)")
            CP00_imputer = SimpleImputer(strategy='mean')

        # Apply the CP00_imputer
        CP00_df[[col]] = CP00_imputer.fit_transform(CP00_df[[col]])

    return CP00_df

# Assuming CP00_sample_data_imputed and CP00_non_categorical_columns are defined
CP00_sample_data_imputed = CP00_impute_columns_with_mean_or_median(CP00_sample_data_imputed, CP00_non_categorical_columns)
````

## Transformation

````python
import numpy as np
from scipy.stats import boxcox
from sklearn.preprocessing import PowerTransformer
import pandas as pd

# Function to handle transformations based on distribution characteristics
def apply_transformations(CP00_sample_data, columns):
    # Initialize a list to store transformation logs
    CP00_transformation_logs = []

    for column in columns:
        # Compute CP00_skewness and CP00_kurtosis
        CP00_skewness = CP00_sample_data[column].skew()
        CP00_kurtosis = CP00_sample_data[column].kurt()
        CP00_action = "None"  # Default CP00_action

        # Handle Right Skew (Positive Skew)
        if CP00_skewness > 1:
            CP00_action = "Log Transformation"
            CP00_sample_data[column] = np.log1p(CP00_sample_data[column])

        # Handle Left Skew (Negative Skew)
        elif CP00_skewness < -1:
            CP00_action = "Reflect and Log Transformation"
            CP00_sample_data[column] = np.log1p(CP00_sample_data[column].max() - CP00_sample_data[column])

        # Handle High Kurtosis (Heavy Tails)
        if CP00_kurtosis > 3:
            try:
                CP00_action = "Box-Cox Transformation"
                CP00_sample_data[column], _ = boxcox(CP00_sample_data[column].clip(lower=1))
            except ValueError:
                CP00_action = "Box-Cox Failed, Applied Yeo-Johnson"
                transformer = PowerTransformer(method='yeo-johnson')
                CP00_sample_data[column] = transformer.fit_transform(CP00_sample_data[[column]])

        # Handle Low Kurtosis (Light Tails)
        elif CP00_kurtosis < 3 and CP00_action == "None":
            CP00_action = "Yeo-Johnson Transformation"
            transformer = PowerTransformer(method='yeo-johnson')
            CP00_sample_data[column] = transformer.fit_transform(CP00_sample_data[[column]])

        # Append the log entry
        CP00_transformation_logs.append({
            'Column Name': column,
            'Skewness': CP00_skewness,
            'Kurtosis': CP00_kurtosis,
            'Action Taken': CP00_action
        })

    # Create a DataFrame for transformation logs
    transformation_log_CP00_df = pd.DataFrame(CP00_transformation_logs)
    return CP00_sample_data, transformation_log_CP00_df

# Example usage with CP00_sample_data_encoded
CP00_sample_data_encoded = CP00_sample_data_ordinally_encoded.copy()
CP00_sample_data_encoded, CP00_transformation_logs = apply_transformations(CP00_sample_data_encoded, CP00_non_categorical_columns)

# Display the transformation log DataFrame
print("Transformation Log:")
print(CP00_transformation_logs)
````

## KMeans Clustor Analysis

````python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import OrdinalEncoder

# Copy of the encoded DataFrame for clustering
CP00_sample_data_cluster = CP00_sample_data_encoded.copy()

# Ensure all categorical columns are numeric using OrdinalEncoder
encoder = OrdinalEncoder()
CP00_sample_data_cluster[CP00_categorical_columns] = encoder.fit_transform(
    CP00_sample_data_cluster[CP00_categorical_columns].astype(str)
)

# Ensure all columns are numeric
clustering_columns = CP00_categorical_columns + CP00_non_categorical_columns
for col in clustering_columns:
    CP00_sample_data_cluster[col] = pd.to_numeric(CP00_sample_data_cluster[col], errors='coerce')

# Drop rows with any NaN values
CP00_sample_data_cluster.dropna(subset=clustering_columns, inplace=True)

# Adjust k values based on dataset size
max_clusters = min(len(CP00_sample_data_cluster) - 1, 6)  # Ensure valid cluster range
CP00_k_values = range(2, max_clusters + 1)

# Initialize lists to store silhouette scores
CP00_silhouette_scores = []

# Iterate over the range of cluster numbers
for CP00_k in CP00_k_values:
    # Initialize KMeans with the current number of clusters
    CP00_kmeans = KMeans(n_clusters=CP00_k, random_state=55002, n_init=10)

    # Fit the model and predict cluster labels
    CP00_cluster_labels = CP00_kmeans.fit_predict(CP00_sample_data_cluster[clustering_columns])

    # Calculate the silhouette score and append it to the list
    CP00_silhouette_scores.append(
        silhouette_score(CP00_sample_data_cluster[clustering_columns], CP00_cluster_labels)
    )

# Display silhouette scores
for k, score in zip(CP00_k_values, CP00_silhouette_scores):
    print(f"Silhouette score for {k} clusters: {score:.2f}")
````

## Elbow Plot

````python
import matplotlib.pyplot as plt

# Enhanced Silhouette Score Plot
plt.figure(figsize=(12, 8))  # Increased figure size for better visibility
plt.plot(CP00_k_values, CP00_silhouette_scores, marker='o', linestyle='--', color='b', linewidth=2)
plt.scatter(CP00_k_values, CP00_silhouette_scores, color='red', label='Silhouette Points')  # Highlight individual points

# Titles and Labels
plt.title('Silhouette Scores for Optimal Cluster Determination', fontsize=16, fontweight='bold')
plt.xlabel('Number of Clusters (k)', fontsize=14)
plt.ylabel('Silhouette Score', fontsize=14)

# Adding annotations for each score
for i, score in enumerate(CP00_silhouette_scores):
    plt.text(CP00_k_values[i], score, f'{score:.2f}', fontsize=10, ha='center', va='bottom')

# Grid, Legend, and Styling
plt.grid(visible=True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12, loc='best')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Show the plot
plt.show()
````
